```python
import datetime
import schedule
import time
import matplotlib.pyplot as plt
```

```python
import matplotlib.pyplot as plt

# TEST 
labels = ['', 'SK', 'LG', '', '']
ratio = [50, 20, 10, 10, 10]


def GoogleTimer():
  plt.pie(ratio, labels=labels, shadow=True, startangle=90)
  plt.show()
  

# 1 GoogleTimer 
schedule.every(1).minutes.do(GoogleTimer)

#  

while True:
  schedule.run_pending()
  time.sleep(1)
  print(datetime.datetime.now())
```
