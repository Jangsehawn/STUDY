```python
import datetime
import schedule
import time
import matplotlib.pyplot as plt
```

```python
import matplotlib.pyplot as plt

# TEST 
labels = ['삼성전자', 'SK하이닉스', 'LG전자', '네이버', '카카오']
ratio = [50, 20, 10, 10, 10]


def GoogleTimer():
  plt.pie(ratio, labels=labels, shadow=True, startangle=90)
  plt.show()
  

# 1분마다 GoogleTimer를 반복
schedule.every(1).minutes.do(GoogleTimer)

# 실행 코드

while True:
  schedule.run_pending()
  time.sleep(1)
  print(datetime.datetime.now())
```

## 반복수행을 하는데 matplot은 수행시간이 너무 긴 것 같아서 plotly를 사용해보았다 

```python
# matplotlib 작업시간
start = time.time() 
plt.pie(ratio, labels=labels, shadow=True, startangle=90)
print("time :", time.time() - start)
```

```python
import plotly.offline
import plotly.express as px
import pandas as pd



test_plotly = pd.DataFrame({'labels':['삼성전자', 'SK하이닉스', 'LG전자', '네이버', '카카오'],'ratio':[50, 20, 10, 10, 10]})
```

```python
# plotly 작업시간
start = time.time() 
fig = px.pie(test_plotly, values='ratio')
plotly.offline.plot(fig)
fig.show()
print("time :", time.time() - start)
```
